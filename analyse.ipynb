```python
import csv
import numpy as np

import matplotlib.pyplot as plt

num = []
time = []


with open('./complexity.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        num.append(row[0])
        time.append(float(row[1]))
```

```python
def moving_average(data, window_size):
    if len(data) < window_size:
        window_size = len(data)  # Ajuste dynamiquement la fenêtre
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

# Paramètres
window_size = 110
linwidth = 1

# Calcul de la moyenne glissante pour time_basic
y_smooth = moving_average(time, window_size)
x = num[:len(y_smooth)]

# Tracer les résultats
plt.figure(figsize=(50, 10))
plt.plot(x, y_smooth, linewidth=linwidth, label='SEARCH PATTERN TIME COMPLEXITY')

plt.show()
```
